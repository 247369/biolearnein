# Manejo de Datos con Pandas

Pandas es una biblioteca utilizada para la manipulación y análisis de datos tabulares, similar a las hojas de cálculo.


```python
import pandas as pd
import seaborn as sns
# Manejo de tablas y organización de datos con pandas
df = pd.read_csv('https://raw.githubusercontent.com/LuisPerezTimana/Webinars/main/diabetes.csv')
df
```

```python
# Selección de columnas relevantes
df = df.loc[:, ["Glucose", "BMI", "Outcome"]]
df
```

```python
# Filtrar datos
filtrado = df[df.BMI > 50]
filtrado
```

```python
# Estadísticas descriptivas
df.describe()
```

# Visualización de Datos

 En esta sección utilizamos Matplotlib para graficar y analizar relaciones entre variables


```python
import matplotlib.pyplot as plt
# Gráficos básicos con matplotlib

# Filtrar datos por categoría según el valor de Outcome
# D: Diabético (Outcome = 1)
# C: Control (Outcome = 0)

D = df[df.Outcome == 1 ]
C = df[df.Outcome == 0 ]

# Crear un gráfico de dispersión
plt.figure()

plt.scatter(C.Glucose, C.BMI, label = "Control")
plt.scatter(D.Glucose, D.BMI, label = "Diabetes")

plt.legend()
plt.show()
```

# Preprocesamiento y Entrenamiento de Modelos

Para construir un modelo predictivo, primero debemos preprocesar los datos y luego dividirlos en conjuntos de entrenamiento y prueba.


```python
# Separar en x e y
# x: todas las variables, menos la de salida
# y: variable de salida

x = df.iloc[:, :-1]
y = df.iloc[:, -1]
x
```

# Estandarizar datos

Esto asegura que todas las variables tengan una escala similar


```python
from sklearn.preprocessing import StandardScaler, Normalizer

# Estandarización de datos
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
x_scaled

```

# Dividir datos en entrenamiento y prueba

El conjunto de entrenamiento se usa para ajustar el modelo, y el de prueba para evaluarlo

```python
from sklearn.model_selection import train_test_split

# División de datos en conjuntos de entrenamiento y testeo
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.3, random_state = 42)
print(x_train.shape)
```

```python
# Mostrar data
y_test
```

# Entrenar modelo

Random Forest es un algoritmo de clasificación basado en múltiples árboles de decisión

```python
from sklearn.ensemble import RandomForestClassifier
# Random Forest

#rf = RandomForestClassifier(max_depth = 3)
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
from sklearn import tree
# Visualizamos el primer árbol del bosque aleatorio
plt.figure()

tree.plot_tree(rf.estimators_[0], filled = True)

plt.show()

```

# Predicción y evaluación

```python
# Predicción de la data de prueba
y_pred_rf = rf.predict(x_test)
y_pred_rf
```

```python
# Predicción de la data de prueba en probabilidades
y_pred_prob_rf = rf.predict_proba(x_test)
y_pred_prob_rf
```

En esta sección se evalúa el modelo mediante métricas como la exactitud y la curva ROC.


```python
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, confusion_matrix
# Calcular la exactitud del modelo
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_rf

```

# Curvas ROC

roc-curve-v2.webp

La curva ROC mide el rendimiento del modelo en clasificación binaria

```python
# Curvas ROC
# La curva ROC mide el rendimiento del modelo en clasificación binaria
fpr, tpr, _ = roc_curve(y_test, y_pred_prob_rf[:,1])
area = roc_auc_score(y_test, y_pred_prob_rf[:,1])

print(area)

plt.figure()

plt.plot(fpr, tpr, label = "RF AUC: " + str(round(area, 2)))
plt.plot([0, 1], [0, 1], "k--")


plt.legend()
plt.show()
```

# Matrices de confusión

Esta matriz muestra el número de predicciones correctas e incorrectas para cada clase

image (2).png

```python
# Matrices de confusión
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize = (10, 5))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicción')
plt.ylabel('Actual')
plt.title('Matriz de Confusión - Random Forest')

```

# Límite de Decision

Esta sección grafica cómo el modelo clasifica las observaciones en el espacio de características

pasted image 0.png

```python
# Visualización de la predicción en el espacio de características
x_min, x_max = x.Glucose.min() - 1, x.Glucose.max() + 1
y_min, y_max = x.BMI.min() - 1, x.BMI.max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1), np.arange(y_min, y_max, 1))
matriz = scaler.fit_transform(np.c_[xx.ravel(), yy.ravel()])
Z = rf.predict(matriz)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(C.Glucose, C.BMI, label = "Control")
plt.scatter(D.Glucose, D.BMI, label = "Diabético")
plt.legend()
plt.title('Predicción con Random Forest')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```
